```python
tmax_cdf = xr.open_dataset(r"G:\popoes\Projects\Juncal_norte\Climate_forcing\CR2MET_TMAX_JUNCALNORTE_500m.nc")

tmax = rio.open(r"G:\popoes\Projects\Juncal_norte\Climate_forcing\CR2MET_TMAX_JUNCALNORTE_500m.nc")
tmin = rio.open(r"G:\popoes\Projects\Juncal_norte\Climate_forcing\CR2MET_TMIN_JUNCALNORTE_500m.nc")

tmax_50 = pd.read_excel(r"C:\Users\pauls\Desktop\Proyectos\Juncal_norte\Scenarios\tmax_T50_2.xlsx")
tmax_100 = pd.read_excel(r"C:\Users\pauls\Desktop\Proyectos\Juncal_norte\Scenarios\tmax_T100_2.xlsx")

tmin_50 = pd.read_excel(r"C:\Users\pauls\Desktop\Proyectos\Juncal_norte\Scenarios\tmin_T50_2.xlsx")
tmin_100 = pd.read_excel(r"C:\Users\pauls\Desktop\Proyectos\Juncal_norte\Scenarios\tmin_T100_2.xlsx")

#Read net dem
netdem_t = rio.open(r"C:\Users\pauls\Desktop\Proyectos\Juncal_norte\GIS\net_dem.tif")
netdem = netdem_t.read(1)

#Read mask
mask_tif = rio.open(r"C:\Users\pauls\Desktop\Proyectos\Juncal_norte\GIS\juncalnorte_mask_v0.tif") #windows
mask_array = mask_tif.read(1)

#Set gradient
g_tmp = -0.0067 #PP gradient mm/m related to pp_avg

for i in range(len(tmax_50)):
    ii = i+1
    if ii < 10000:
        value = ii/1000
        f_value = "{:.3f}".format(value)
        #t50
        name1 = r"C:\Users\pauls\Desktop\Proyectos\Juncal_norte\Scenarios\tmax8\tmax800"+str(f_value)
        name2 = r"C:\Users\pauls\Desktop\Proyectos\Juncal_norte\Scenarios\tmin8\tmin800"+str(f_value)
        name3 = r"C:\Users\pauls\Desktop\Proyectos\Juncal_norte\Scenarios\tavg8\tavg800"+str(f_value)
        #t100
        name4 = r"C:\Users\pauls\Desktop\Proyectos\Juncal_norte\Scenarios\tmax9\tmax900"+str(f_value)
        name5 = r"C:\Users\pauls\Desktop\Proyectos\Juncal_norte\Scenarios\tmin9\tmin900"+str(f_value)
        name6 = r"C:\Users\pauls\Desktop\Proyectos\Juncal_norte\Scenarios\tavg9\tavg900"+str(f_value)
      
    tmax_it1 = tmax.read(ii)
    tmin_it1 = tmin.read(ii)
    
    tmax_it2 = tmax.read(ii)
    tmin_it2 = tmin.read(ii)

    #Elevation adjustment
    tmax_it1 = tmax_it1*0+tmax_50.iloc[i,1]+g_tmp*(netdem+14)
    tmin_it1 = tmin_it1*0+tmin_50.iloc[i,1]+g_tmp*(netdem+14)
    tavg_it1 = (tmax_it1+tmin_it1)/2

    tmax_it2 = tmax_it2*0+tmax_100.iloc[i,1]+g_tmp*(netdem+14)
    tmin_it2 = tmin_it2*0+tmin_100.iloc[i,1]+g_tmp*(netdem+14)
    tavg_it2 = (tmax_it2+tmin_it2)/2

    #### t50
    #Tmax
    maps1 = rio.open(
    str(name1),"w",
    driver = "PCRaster",
    pcraster_valuescale = "VS_SCALAR",
    height = mask_tif.shape[0],
    width = mask_tif.shape[1],
    count = 1,
    nodata = mask_tif.nodata,
    dtype = np.float32,
    crs = 32719,
    transform = mask_tif.transform)
    maps1.write(tmax_it1,1)
    maps1.close()
    
    #Tmin
    maps2 = rio.open(
    str(name2),"w",
    driver = "PCRaster",
    pcraster_valuescale = "VS_SCALAR",
    height = mask_tif.shape[0],
    width = mask_tif.shape[1],
    count = 1,
    nodata = mask_tif.nodata,
    dtype = np.float32,
    crs = 32719,
    transform = mask_tif.transform)
    maps2.write(tmin_it1,1)
    maps2.close()
    
    #Tavg
    maps3 = rio.open(
    str(name3),"w",
    driver = "PCRaster",
    pcraster_valuescale = "VS_SCALAR",
    height = mask_tif.shape[0],
    width = mask_tif.shape[1],
    count = 1,
    nodata = mask_tif.nodata,
    dtype = np.float32,
    crs = 32719,
    transform = mask_tif.transform)
    maps3.write(tavg_it1,1)
    maps3.close()

    print(name1)
    print(name2)
    print(name3)
    
    
    #### t100
    #Tmax
    maps4 = rio.open(
    str(name4),"w",
    driver = "PCRaster",
    pcraster_valuescale = "VS_SCALAR",
    height = mask_tif.shape[0],
    width = mask_tif.shape[1],
    count = 1,
    nodata = mask_tif.nodata,
    dtype = np.float32,
    crs = 32719,
    transform = mask_tif.transform)
    maps4.write(tmax_it2,1)
    maps4.close()
    
    #Tmin
    maps5 = rio.open(
    str(name5),"w",
    driver = "PCRaster",
    pcraster_valuescale = "VS_SCALAR",
    height = mask_tif.shape[0],
    width = mask_tif.shape[1],
    count = 1,
    nodata = mask_tif.nodata,
    dtype = np.float32,
    crs = 32719,
    transform = mask_tif.transform)
    maps5.write(tmin_it2,1)
    maps5.close()
    
    #Tavg
    maps6 = rio.open(
    str(name6),"w",
    driver = "PCRaster",
    pcraster_valuescale = "VS_SCALAR",
    height = mask_tif.shape[0],
    width = mask_tif.shape[1],
    count = 1,
    nodata = mask_tif.nodata,
    dtype = np.float32,
    crs = 32719,
    transform = mask_tif.transform)
    maps6.write(tavg_it2,1)
    maps6.close()

    print(name4)
    print(name5)
    print(name6)
```
